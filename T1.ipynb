# Lendo Primeira linha -> Entrada de TESTE

```r
info_obj <- function(obj) {
    # Função para exibir informações do objeto
    classe <- class(obj)
    tipo <- typeof(obj)
    tam <- object.size(obj)

    cat(obj, classe, tipo, tam, "bytes\n")
}
```

```r
# Lendo a Primeira Linha
#file <- "/content/entrada_testes.txt"
file <- "/content/entrada.txt"
primeira_linha <- strsplit(
    readLines(file, n = 1), " "
    )[[1]]

# Serarando e Convertendo Inteiro Quantidade de Meses
m <- primeira_linha[1] # Qtd Meses Contratação
m <- as.integer(m)

# Separando e Convertendo Inteiro Fornecedores
n <- primeira_linha[2] # Fornecedores
n <- as.integer(n)

# Separando e Convertendo Doble Taxa
t <- primeira_linha[3] # Taxa
t <- as.double(t)

sprintf(
    "Quantidade Meses %d - Fornecedores %d - Taxa %.2f",
    m, n, t
)

info_obj(m)
info_obj(n)
info_obj(t)
```

# Separando os dados a serem processados

```r
# Lendo Restante do arquivo, são os dados a serem processados.
#dados_uteis <- readLines(file , n = -1) # Entrada de TESTE

dados_uteis <- readLines(con <- file(file))
close(con)

# Ordenando o Vector
dados_uteis <- sort(dados_uteis)

#dados_uteis <- dados_uteis[-length(dados_uteis)]
```

```r
array_3D <- array(Inf, dim=c(m,m,n))
#print(array_3D)
cat("Dimensões:", dim(array_3D))
```

```r
for (i in dados_uteis) {
    linha <- as.numeric(strsplit(i, " ")[[1]])

    if (length(linha) == 4) {
        #print("Tamanho 4")
        array_3D[linha[2], linha[3], linha[1]] <- linha[4]
    }else {
        print("Retirando Linha 1")
    }
}

print(array_3D)
```
